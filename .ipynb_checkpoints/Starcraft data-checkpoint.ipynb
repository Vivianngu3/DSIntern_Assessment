```python
# Indentifying the number of rows in each LeagueIndex
# There is a gap collected between higher and lower ranks

def ranked_rows (rank):
    
    Ranks = df[df['LeagueIndex']== rank]
    print(Ranks.shape)

print(ranked_rows(1), ranked_rows(2), ranked_rows(3), ranked_rows(4),
      ranked_rows(5), ranked_rows(6), ranked_rows(7), ranked_rows(8))
```

```python
# Double checking for nulls in Columns
# The nulls were placed as question marks
rows_with_question = df[df['TotalHours'] == '?']
rows_with_question
```

```python
# Removing the columns with nulls for all of rank 8 (Collecting hours would be important)
Remove_columns = ['Age', 'TotalHours', 'HoursPerWeek', ]
New_df = df.drop(columns=Remove_columns)
```

```python
# Finding summary stats
New_df.describe()
```

```python
#Function to find and remove the outliers
Test = New_df[['APM', 'SelectByHotkeys', 'AssignToHotkeys', 'UniqueHotkeys', 'MinimapAttacks', 'MinimapRightClicks', 'NumberOfPACs', 
        'GapBetweenPACs', 'ActionLatency', 'ActionsInPAC', 'TotalMapExplored', 'WorkersMade', 'UniqueUnitsMade', 
        'ComplexUnitsMade', 'ComplexAbilitiesUsed']]


def outlier_removal(Test):
    
# Upper and lower limits
    upper_limit = Test.mean() + 3 * Test.std()
    lower_limit = Test.mean() - 3 * Test.std()

    print("Upper limit:", upper_limit)
    print("Lower limit:", lower_limit)

# Filter based on limits
    cleaned_data = Test[((Test >= lower_limit) & (Test <= upper_limit)).all(axis=1)]
    return cleaned_data

cleaned_data = outlier_removal(Test)

cleaned_data
```

```python
#Adding back LeagueIndex and GameID

label_column = New_df.loc[cleaned_data.index]['LeagueIndex']
gamerId = New_df.loc[cleaned_data.index]['GameID']

cleaned_df = pd.concat([cleaned_data, label_column, gamerId], axis=1)

cleaned_df
```

```python
# Showing the distribution of data across LeagueIndex after outlier removal
def ranked (rank):
    
    Ranks = cleaned_df[cleaned_df['LeagueIndex']== rank]
    print(Ranks.shape)

print(ranked(1), ranked(2), ranked(3), ranked(4),
      ranked(5), ranked(6), ranked(7), ranked(8))
```

```python
# Box plot to visualize removed outliers
plt.figure(figsize=(12, 6))
cleaned_df[cols].boxplot()

plt.title('Box Plot of Starcraft Player Performance')
plt.xlabel('Columns')
plt.ylabel('Values')

plt.xticks(rotation=90)  #Adjusts the rotation angle of column label to be more readable

# Adjust the layout to prevent label overlapping
plt.tight_layout()

# Display the plot
plt.show()
```
